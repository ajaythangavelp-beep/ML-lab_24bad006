```python
df = pd.read_csv("data.csv", encoding='latin-1')
```

```python
df.head()
df.tail()
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
product_sales = df.groupby("Description")["Quantity"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
product_sales.plot(kind="bar")
plt.title("Top-Selling Products")
plt.ylabel("Total Quantity Sold")
plt.show()
```

```python
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])
daily_sales = df.groupby(df["InvoiceDate"].dt.date)["Quantity"].sum()

plt.figure(figsize=(14,6))
plt.plot(daily_sales)
plt.title("Daily Sales Trend")
plt.xlabel("Date")
plt.ylabel("Quantity Sold")
plt.show()
```

```python
df = pd.read_csv("/content/Churn_Modelling.csv")
```

```python
# AGE - Bar Plot
plt.figure(figsize=(12,5))
df["Age"].value_counts().sort_index().plot(kind="bar")
plt.title("Age Distribution  Bar Plot")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

# AGE - Box Plot
plt.figure(figsize=(8,4))
sns.boxplot(x=df["Age"])
plt.title("Age Distribution  Boxplot")
plt.show()

# INCOME (EstimatedSalary) - Bar Plot
plt.figure(figsize=(12,5))
df["EstimatedSalary"].value_counts().sort_index().head(50).plot(kind="bar")
plt.title("Income Distribution  Bar Plot")
plt.xlabel("Estimated Salary")
plt.ylabel("Count")
plt.show()

# INCOME - Box Plot
plt.figure(figsize=(8,4))
sns.boxplot(x=df["EstimatedSalary"])
plt.title("Income Distribution  Boxplot")
plt.show()

# SPENDING (Balance) - Bar Plot
plt.figure(figsize=(12,5))
df["Balance"].value_counts().sort_index().head(50).plot(kind="bar")
plt.title("Balance Distribution  Bar Plot")
plt.xlabel("Balance Amount")
plt.ylabel("Count")
plt.show()

# SPENDING - Box Plot
plt.figure(figsize=(8,4))
sns.boxplot(x=df["Balance"])
plt.title("Balance Distribution  Boxplot")
plt.show()
```
